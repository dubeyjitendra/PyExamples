```python
## read SQL
from sqlite3 import connect
conn = connect(':memory:')
df = pd.DataFrame(data=[[0, '10/11/12'], [1, '12/11/10']],
                  columns=['int_column', 'date_column'])

df.to_sql('test_data', conn)
```

```python
read_sql_data = pd.read_sql('SELECT int_column, date_column FROM test_data', conn)
read_sql_data
```

```python
# parse dates
x =pd.read_sql('SELECT int_column, date_column FROM test_data',
            conn,
            parse_dates=["date_column"])
x
```

```python
y = pd.read_sql('SELECT int_column, date_column FROM test_data',
            conn,
            parse_dates={"date_column": {"format": "%y/%m/%d"}})
y
```

```python
### convert to other format

y.to_sql("name.sql",conn)
```
