## markers

```python
# set the matplotlib inline
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#### check this link to get more 
# https://matplotlib.org/stable/api/markers_api.html
```

```python
x= [12,34,26]
y= [23,45,67]

plt.plot(x,y,marker='o') #
plt.show()
```

### Set marker with markersize

```python
x= [12,34,26]
y= [23,45,67]

plt.plot(x,y,marker='o',markersize=15) #
plt.show()
```

### Set marker with markeredgewidth(width outer area)

```python
x= [12,34,26]
y= [23,45,67]

plt.plot(x,y,marker='o',markersize=15,markeredgewidth=10) #
plt.show()
```

### Set marker with markeredgewidth(width outer area) with color

```python
x= [12,34,26]
y= [23,45,67]

plt.plot(x,y,marker='o',markersize=15,markeredgewidth=2,markeredgecolor='red') #
plt.show()
```

### Set marker with markeredgewidth(width outer area) with facecolor

```python
x= [12,34,26]
y= [23,45,67]

plt.plot(x,y,marker='o',markersize=15,markeredgewidth=2,markeredgecolor='red',markerfacecolor='green') #
plt.show()
```
