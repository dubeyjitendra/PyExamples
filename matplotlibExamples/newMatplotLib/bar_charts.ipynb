### bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
x = np.arange(len(company))
x
```

```python
plt.bar(x,revenue, label="Revenue")
plt.xticks(x,company)
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(x-0.2,revenue,  width=0.4,  label="Revenue")
plt.bar(x+0.2,profit, width=0.4, label="Profit", color="orange")
plt.legend()
```

```python
plt.barh(x,revenue, label="Revenue")
```

```python
plt.barh(x+0.2,revenue, height=0.4,  label="Revenue")
plt.barh(x-0.2,profit,height=0.4,  label="Profit", color="orange")
plt.legend()
```
