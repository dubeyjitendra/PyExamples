### Display Multiple Plots

subplots() function you can draw multiple plots in one place or figure

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

##### 
The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

### Without Numpy 

```python
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]


plt.subplot(1,2,2) #row,colums and size
plt.plot(x,y, label='x')

x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

plt.subplot(1,2,2) # row,colums and size 
plt.plot(x,y, label='y')
plt.legend()
plt.show()
```

### With NumPy 

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # #the figure has 1 row, 2 columns, and this plot is the first plot. even you can change the row and column value
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # #the figure has 1 row, 2 columns, and this plot is the second plot.even you can change the row and column
plt.plot(x,y)

plt.show()
```

### Add title and super title

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # #the figure has 1 row, 2 columns, and this plot is the first plot. even you can change the row and column value
plt.plot(x,y)
plt.title("Diamond")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # #the figure has 1 row, 2 columns, and this plot is the second plot.even you can change the row and column
plt.plot(x,y)
plt.title("Gold")
plt.suptitle("Market")
plt.show()
```
