# Matplotlib Overview

```python
# import packages
%matplotlib inline
from matplotlib import pyplot as plt 
```

```python
plt.plot([2,6,8],[8,12,24])
plt.show()
```

### Change linewidth

```python
plt.plot([2,6,8],[8,12,24], linewidth=10)
plt.show()
```

# Set the Title Name

```python
plt.plot([2,6,8],[8,12,24])
plt.title("First plot")
plt.show()
```

### Set xlabel and ylablel

```python
plt.plot([2,6,8],[8,12,24])
plt.title("First plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

## Grid

```python
plt.plot([2,6,8],[8,12,24])
plt.title("First plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

## Grid X axis

```python
plt.plot([2,6,8],[8,12,24])
plt.title("First plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(axis='x')
plt.show()
```

### Grid Y axis

```python
plt.plot([2,6,8],[8,12,24])
plt.title("First plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(axis='y')
plt.show()
```

## Set STyle 

```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
x= [2,5,6]
y= [8,20,24]

style.use('ggplot')
plt.plot(x,y)
plt.show()
```

## Add legend

```python
x= [2,5,6]
y= [8,20,24]

style.use('ggplot')
plt.plot(x,y)
plt.legend("X axis")
plt.show()
```

### Compare two Data Points

```python
from matplotlib import pyplot as plt
```

```python
x1= [2,4,6,8,10]
y1=[3,6,9,12,15]

x2 =[4,8,12,16,20]
y2=[5,10,15,20,25]

plt.plot(x1,y1,label="First")
plt.plot(x2,y2,label="Second")
# adding legends before define then no need to set inside the legend
plt.legend()
plt.show()
```

### Set Scatter chart

```python
x1= [2,4,6,8,10]
y1=[3,6,9,12,15]
plt.scatter(x1,y1,label="Set Scatter")
plt.show()
```

```python
x1= [2,4,6,8,10]
y1=[3,6,9,12,15]
x2 =[4,8,12,16,20]
y2=[5,10,15,20,25]
plt.scatter(x1,y1,label="First Scatter")
plt.scatter(x2,y2,label="second Scatter")
plt.legend()
plt.show()
```
